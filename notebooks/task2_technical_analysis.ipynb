```python
import sys
import os

# Add src folder to path
sys.path.append(os.path.abspath("../src"))

from technical_analysis import load_stock_data, apply_technical_indicators, plot_price_with_indicators
```

```python
data_folder = "../data/yfinance_data"
tickers = [f.split("_")[0] for f in os.listdir(data_folder) if f.endswith(".csv")]

for ticker in tickers:
    path = os.path.join(data_folder, f"{ticker}_historical_data.csv")
    df = load_stock_data(path)
    df = apply_technical_indicators(df)
    plot_price_with_indicators(df, ticker)
```
